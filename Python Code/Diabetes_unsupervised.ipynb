```python
x = data2.loc[:,[0,2]].values
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init = 'k-means++', random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('No Of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters= 3, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'Cluster 3')
#plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1],s = 100, c = 'cyan', label = 'Cluster 4')
#plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1],s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('BloodPressure (k$)')
plt.ylabel('Age (1-100)')
plt.legend()
plt.show()
```

```python
x = data2[[0,1,2,3,4,5,6,7]]
y = data1[['Diabetes']]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
x_std=StandardScaler().fit_transform(x)
```

```python
x_std
```

```python
features=x_std.T
covariance_matrix=np.cov(features)
print(covariance_matrix)
```

```python
eig_vals,eig_vecs=np.linalg.eig(covariance_matrix)
```

```python
print('\nEigen values \n%s' %eig_vals)
```

```python
eig_vals[0]/sum(eig_vals)
```

```python
projected_x=x_std.dot(eig_vecs.T[0])
```

```python
projected_x
```

```python
result=pd.DataFrame(projected_x,columns=['PC1'])   # PC1=Principal Component 1
result['y-axis']=0.0
result['label']=y
result
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.lmplot('PC1','y-axis',data=result,fit_reg=False,scatter_kws={'s':50},hue='label')

#title
plt.title('PCA result')

# hue='label' .....unknown labels
# fit_reg.... fit regression
```
